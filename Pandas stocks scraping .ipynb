```python
import pandas as pd
import pandas_datareader as web
import mplfinance as mpf
from datetime import date

class Stock_analysis():
    def __init__(self):
        self.code = []
        print("To use this program, if you are want to check for Korea stock, please choose Naver." 
              "If you are checking US stock, select yahoo.\n\n")
        stock = input("Give me your code name:   1. naver, 2. yahoo")
        
        if stock == 1: self.type = "naver"
        else: self.type = "yahoo"
        self.type_data()
    
    def type_data(self, start="2018-01-01", end=str(date.today())):
        while True:
            ask = input("Give me your code name:      ").upper()
            if ask != 'END':
                self.code.append(ask)
            else:
                break

        if self.code:
            for code in self.code:
                try:
                    df = web.DataReader(code, data_source=self.type, start=start, end=end)
                    mpf.plot(df)
                
                except:
                    print("Wrong code:  {}".format(code))
                    
stock = Stock_analysis()
```
